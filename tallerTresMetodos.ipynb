```python
tablaParaX = []
for i in enumerate(historialNewton, start=1):
    tablaParaX.append(f"X{i}")

print(tablaParaX)

import matplotlib.pyplot as plt

x_indices = list(range(1, len(historialNewton) + 1))
plt.figure()
plt.plot(x_indices, historialNewton, marker='o')
plt.xlabel('Iteración')
plt.ylabel('Valor de x')
plt.title('Historial de Newton')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
##Graficar tangentes 
import numpy as np 
import matplotlib as plt 
from scipy.optimize import newton 
historialNewton = [ ]
def f(x):
    y=x**3 -3*x**2+x-1
    print(x,y)
    historialNewton.append(x)
    return y

def df(x):
    yPrim=3*x**2-6*x+1
    return yPrim

raiz = newton(f, x0=100, fprime=df, maxiter=50)
print(historialNewton)


import matplotlib.pyplot as plt

historialNewton = []
def f(x):
    y = x**3 - 3*x**2 + x - 1
    print(x, y)
    historialNewton.append(x)
    return y

def df(x):
    return 3*x**2 - 6*x + 1

raiz = newton(f, x0=100, fprime=df, maxiter=50)
print(historialNewton)
print("Raíz encontrada:", raiz)

x_min, x_max = 0, 6
x_curve = np.linspace(x_min, x_max, 300)
y_curve = x_curve**3 - 3*x_curve**2 + x_curve - 1

plt.figure(figsize=(10, 6))
plt.plot(x_curve, y_curve, color="black", label="f(x)")

for idx, xi in enumerate(historialNewton):
    yi = xi**3 - 3*xi**2 + xi - 1
    slope = 3*xi**2 - 6*xi + 1
    x_tangent = np.linspace(x_min, x_max, 200)
    y_tangent = yi + slope * (x_tangent - xi)
    label = "Tangente" if idx == 0 else None
    plt.plot(x_tangent, y_tangent, color="orange", alpha=0.25, label=label)

y_points = [xi**3 - 3*xi**2 + xi - 1 for xi in historialNewton]
plt.scatter(historialNewton, y_points, color="red", zorder=5, label="Iteraciones")
plt.scatter([raiz], [0], color="blue", zorder=6, label=f"Raíz ~ {raiz:.6f}")

plt.xlim(x_min, x_max)
plt.ylim(-10, 20)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Tangentes en cada punto del historial de Newton (zoom con límite)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
